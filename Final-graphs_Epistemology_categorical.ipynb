```python

def three_figures_loss(x_0, y_0,
                  x_1, y_1,
                  x_2, y_2, vy_2, 
                  filepath):
    
    plt.style.use(['science', 'ieee', 'no-latex'])
    plt.rcParams['font.family'] = "Arial"
    rc('font', weight='normal')

    # Create a figure with three subplots in a row
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (9,2)) #, figsize=(12, 4)

    # Plot the first subplot on the left
    axes[0].scatter(x_0, y_0, color = '#A20346', marker = 's', s= 10)
    axes[0].plot(x_0, y_0, color = '#A20346', marker = 's', label = 'Loss + Validation loss')
    # axes[0].scatter(x_0, vy_0, color = '#FC5A50', marker = 's', s= 10)
    # axes[0].plot(x_0, vy_0, color = '#FC5A50', marker = 's', label = 'Validation')
    #axes[0].errorbar(x_0, y_0, yerr_0, color='#A20346', fmt='none')
    #axes[0].plot(x_0, y_0, linestyle='solid', color='#A20346', alpha=0.5)
    axes[0].set_xlabel('Training ratio', fontsize=10)
    axes[0].set_ylabel('Loss', fontsize=10)
    axes[0].xaxis.set_ticks(np.arange(0.3, 1.0, 0.1))
    axes[0].yaxis.set_ticks(np.arange(0, 2.0, 0.2))
    axes[0].minorticks_off()
    legend = axes[0].legend(framealpha=1, frameon=True, edgecolor = 'black')
    legend.get_frame().set_linewidth(0.2)

# Plot the second subplot in the middle
    axes[1].scatter(x_1, y_1, color = '#004E7E', marker = 's', s= 10,)
    axes[1].plot(x_1, y_1, color = '#004E7E', marker = 's', label = 'Loss + Validation loss')
    # axes[1].scatter(x_1, vy_1, color = '#069AF3', marker = 's', s= 10,)
    # axes[1].plot(x_1, vy_1, color = '#069AF3', marker = 's', label = 'loss')
    #axes[1].errorbar(x_1, y_1, yerr_1, color='#004E7E', fmt='none')
    axes[1].plot(x_1, y_1, linestyle='solid', color='#004E7E', alpha=0.5)
    axes[1].set_xlabel('Training ratio', fontsize=10)
    axes[1].set_ylabel('Loss', fontsize=10)
    axes[1].xaxis.set_ticks(np.arange(0.3, 1.0, 0.1))
    axes[1].yaxis.set_ticks(np.arange(0, 2.0, 0.2))
    axes[1].minorticks_off()
    legend = axes[1].legend( framealpha=1, frameon=True, edgecolor = 'black')
    legend.get_frame().set_linewidth(0.2)

    # Plot the third subplot on the right
    axes[2].scatter(x_2, y_2, color = '#3D4E1D', marker = 's', s= 10)
    axes[2].plot(x_2, y_2, color = '#3D4E1D', marker = 's',label = 'Loss')
    axes[2].scatter(x_2, vy_2, color = '#06C2AC', marker = 's', s= 10)
    axes[2].plot(x_2, vy_2, color = '#06C2AC', marker = 's',label = 'Validation loss')
    #axes[2].errorbar(x_2, y_2, yerr_2, color='#3D4E1D', fmt='none')
    #axes[2].plot(x_2, y_2, linestyle='solid', color='#3D4E1D', alpha=0.5)
    axes[2].set_xlabel('Training ratio', fontsize=10) # , fontweight = 'bold'
    axes[2].set_ylabel('Loss', fontsize=10) # fontweight = 'bold'
    axes[2].xaxis.set_ticks(np.arange(0.2, 1.0, 0.1))
    axes[2].yaxis.set_ticks(np.arange(0, 2.0, 0.2))
    axes[2].minorticks_off()
    legend = axes[2].legend( framealpha=1, frameon=True, edgecolor = 'black')
    legend.get_frame().set_linewidth(0.2)


    # Add a shared y-axis label to the leftmost plot
    # fig.text(0.06, 0.5, 'Amplitude', ha='center', va='center', rotation='vertical')
    # Add a shared x-axis label to the bottom plot
    # fig.text(0.5, 0.06, 'Time', ha='center', va='center')

    # Adjust the spacing between the subplots
    plt.subplots_adjust(wspace=0.4)

    # Adjust the spacing between the subplots
    plt.subplots_adjust(wspace=0.4)
    
    plt.savefig(filepath, dpi = 1000)
    
    # Show theplot
    plt.show()

    
   
    return

```
